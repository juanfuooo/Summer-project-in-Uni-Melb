```python
test = np.arange('1950-01-01' , '2022-12-01', dtype='datetime64[M]')
ninosst = data.sst.sel(time = test, lat = np.arange(-4,4,2), lon = np.arange(190,242,2))

ninosst = np.mean(ninosst,axis=1)
ninosst = np.mean(ninosst,axis=1)
```

```python
meanT = ninosst.groupby('time.month').mean()
print(meanT)
```

```python
mT = meanT
for i in range (72):
    mT = np.append(mT,meanT)

mT = np.delete(mT,[-1])
delsst = ninosst - mT

newT = delsst.time


# 3 months running mean
```
