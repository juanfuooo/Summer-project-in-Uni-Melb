```python
import xarray as xr
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
#import wheeler_kiladis_mjo_filter
#import mjoindices.olr_handling as olr
```

```python
ds = pd.read_csv('RMM.csv',
                 header=None,names=['Year', 'Month','Day','RMM1','RMM2','phase','amplitude'])
```

```python

amp = ds.amplitude

amp = amp.values
```

```python
amp[amp == 999 ] = np.nan
amp[amp == 9.999999999999999e+35 ] = np.nan
```

```python
time = pd.date_range(start="19740601",end="20221217",freq="D")
```

```python
plt.figure(figsize=(60,20))
a=plt.plot(time, amp ,linewidth = 1.5 ,color=[0,0,1])
plt.ylim((0,5))
plt.tick_params(labelsize=50)
plt.title('MJO amplitude',fontsize=100)
```
